```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
import warnings

warnings.filterwarnings('ignore')
```

```python
plt.style.available
```

```python
data = pd.DataFrame()
data['X'] = np.arange(0,10,1)
data['Y'] = data['X']**2
```

```python
data
```

```python
plt.style.use("seaborn-talk")
```

```python
sns.lineplot(data.X, data.Y)
```

```python
sns.lmplot("X", "Y", data=data, fit_reg=True)
```

```python
data = pd.read_csv(r'../../../data/vehicles/vehicles.csv') 
```

```python
sns.scatterplot(x="CO2 Emission Grams/Mile", y="Highway MPG", hue="Drivetrain", data=data)
```

```python
sns.relplot(x='CO2 Emission Grams/Mile', y='Highway MPG', hue="Drivetrain", size="Fuel Type", data=data)
```

```python
sns.catplot(x="Highway MPG", y="Drivetrain", hue="Fuel Type", data=data)
```

```python
sns.boxplot(data=data, x="Combined MPG")
```

```python
sns.boxplot(data=data, x="Combined MPG", y="Fuel Type")
```

```python
dataprep = data.loc[(data['Cylinders'] == 4.0 )| (data['Cylinders'] == 6.0 )]
sns.boxplot(data=dataprep, y="Combined MPG", x="Fuel Type", hue="Cylinders");
plt.xticks(rotation=90);
```

```python
sns.violinplot(data=dataprep, y="Combined MPG", x="Cylinders");
```

```python
sns.histplot(data=data, x="Combined MPG", bins=5);
```

```python
sns.distplot(data["Combined MPG"]);
```
