```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
plt.style.available
```

```python
data = pd.DataFrame()
data['X'] = np.arange(0,10,1)
data['Y'] = data['X']**2
```

```python
data
```

```python
plt.style.use("seaborn-talk")
```

```python
sns.lineplot(data.X, data.Y)
```

```python
sns.lmplot("X", "Y", data=data, fit_reg=True)
```
