```python
import matplotlib.pyplot as plt
import math
import numpy as np
```

```python

x1 = np.arange(0,10,1)
x2 = np.linspace(0,5,10)
```

```python
x1
```

```python
x2
```

```python
plt.plot(x1, x1, label='linear')

plt.plot(x1, x2**2, label='quadratic')

plt.xlabel('x label')
plt.ylabel('y label')

plt.title("Linear vs Non Linear Plots")
plt.legend()
plt.show()

```

```python
fig, [ax1, ax2] = plt.subplots(2,1)
ax1.set_title('LINEAR PLOT')
ax1.plot(x1,x2,label='linear' )
ax1.set_xlabel('x label')     
ax1.set_ylabel('y label')
fig.suptitle('This is a somewhat long figure title', fontsize=16)

ax2.set_title('COSINE PLOT')
ax2.plot(x1,np.cos(x1), label='cosine' )
ax2.set_xlabel('x label')     
ax2.set_ylabel('y label')
fig.legend()

plt.show()

```

```python
x = np.random.randn(10)
y = np.random.randn(10)

plt.scatter(x,y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title("Scatter plot on random data")
plt.show()
```

```python
labels = ['Vegetarians', 'Non-Vegetarians', 'Vegans']
plt.style.use('seaborn-colorblind')
sizes = [40, 45, 15]
explode = (0, 0.1, 0)
fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, shadow=True, startangle=80)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


# make data
X, Y = np.meshgrid(np.linspace(-3, 3, 256), np.linspace(-3, 3, 256))
Z = (1 - X/2 + X**5 + Y**3) * np.exp(-X**2 - Y**2)
levels = np.linspace(np.min(Z), np.max(Z), 7)

# plot
fig, ax = plt.subplots()

ax.contour(X, Y, Z, levels=levels)

plt.show()
```
