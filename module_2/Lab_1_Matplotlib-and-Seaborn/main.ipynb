#### Use plt.subplots(nrows=1, ncols=2) to create the plot.

```python
# Method 2
fig, [ax1, ax2] = plt.subplots(nrows=1, ncols=2)
ax1.set_title('LINEAR PLOT')
ax1.plot(x,y,label='linear' )
ax1.set_xlabel('x label')     
ax1.set_ylabel('y label')

ax2.set_title('QUAD PLOT')
ax2.plot(x,z, label='quad' )
ax2.set_xlabel('x label')     
ax2.set_ylabel('y label')
fig.legend()

plt.show
```

#### Use your previous code but now, resize your plot.
**Hint**: Add the figsize argument in plt.subplots().

#### Now you will have to plot both $y=x^2$ and $y=e^x$ in the same plot. Create two subplots and use a normal scale in the first one and a logarithmic scale in the second one.
**Hint**: Use `set_xscale` and `set_yscale`.

```python
# your code here
fig, [ax1, ax2] = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))
ax1.set_title('Normal Scale', fontweight='bold')
ax1.plot(x**2,np.exp(x), "-b", label='same norm')
ax1.set_xlabel('x label')     
ax1.set_ylabel('y label')

ax2.set_title('LOG Scale', fontweight='bold')
ax2.plot(x**2,np.exp(x), "-r", label='same log')
ax2.set_xlabel('x label')     
ax2.set_ylabel('y label')
ax2.set_yscale('log')

fig.legend(loc="upper center")

plt.show
```

#### Now in addition to the summary statistics, we want to have in the same plot the distribution of `Age`. What kind of plot would you use? Plot it. 

```python
ax = sns.histplot(data = titanic['Age'])
labels = [f"{stat}: {val:.2f}" for stat, val in titanic['Age'].describe().items()]
handles = [plt.Line2D([], [], visible=False) for _ in labels]

ax.legend(
    handles,
    labels,
    loc="best",
    handlelength=0,
)
```
