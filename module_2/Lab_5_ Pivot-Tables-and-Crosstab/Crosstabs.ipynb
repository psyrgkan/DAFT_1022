```python
import pandas as pd
import numpy as np
```

### 1. Import the moovies dataset 

# https://www.kaggle.com/carolzhangdc/imdb-5000-movie-dataset

```python
movies = pd.read_csv(r'../../../data/movie_metadata.csv')
movies.columns
```

### 2. Use crosstabs to show the number of moovies per country

```python
pd.crosstab(movies['country'], movies['color'], margins=True).All
```

### 3. Create a crosstab to show the number of films per director and genre splitted by country

```python
pd.crosstab([movies['director_name'], movies['genres']], movies['country'])
```

### 4. Average duration of filmes grouped by color

```python
pd.crosstab(movies['color'], movies['color'], values=movies['duration'], aggfunc="mean", margins=True).All
```

### 5. Average number of likes per director (assuming that total film's likes are counted as a sum of actor's likes)

```python
pd.crosstab(movies['director_name'], movies['color'], values=movies['cast_total_facebook_likes'], aggfunc="mean")
```
